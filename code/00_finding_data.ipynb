```python
# 크롤링
import requests
from bs4 import BeautifulSoup
import time
# 데이터 관련
import random
import numpy as np
import pandas as pd

url_base = 'https://kin.naver.com/userinfo/answerList.naver?u=tBzCEvJdZCZPwYVGTUlXcL%2Fs47ynA6lZ5LzonpmmrP4%3D'
# headers = { header : '입력!입력!입력!입력!입력!' }


qna_link_list = []

for year in range(2021,2026):
    print("==========================")
    print(f"<< {year}년 데이터 수집 >>")

    for page in range(1,100):

        url_rest = f'&year={year}&page={page}'
        url = url_base + url_rest
        
        try: 
            print(f"[ {year}년 {page}페이지 ]")
    
            response = requests.get(url, headers=headers, timeout=10)   # 여기서부터
            response.raise_for_status()
            response.encoding = 'utf-8' # utf-8로 인코딩
        
            bs = BeautifulSoup(response.text, 'html.parser')            # 여기까지는 밖으로 빠져도 될듯
    
            factors = bs.select('#au_board_list tr td.title a')
            links = [factor.attrs['href'] for factor in factors]
            qna_link_list.extend(links)
    
            print(f"[{year}년 페이지 {page}개 완료 (누적: {len(qna_link_list)}개)")

        except Exception as e:
            print(f" !!! 에러: {e} !!!")
            break

        sleep_time = random.uniform(0.7, 1)
        time.sleep(sleep_time)
        
    df = pd.DataFrame(qna_link_list, columns=['url'])
    df.to_csv(f'qna_links_2021-{year}.csv', index=False, encoding='utf-8')

print(qna_link_list)
```

##### Step 2. 각 지식인 Q&A의 질문과 답변 불러오기

```python
# Step1.에서 얻은 link에 mydetail이 밑 코드 실행을 방해하는 요인일 수 있음 >>> mydetail을 detail로 변경하기

df = pd.read_csv('../raw_data/kin.naver.com/qna_links_2021-2025.csv', encoding='utf-8')

df['url'] = df['url'].str.replace('mydetail', 'detail', regex=False)

df.to_csv('../raw_data/kin.naver.com/qna_links_2021-2025_ver2.csv', index=False, encoding='utf-8')
```

```python
# 크롤링
import requests
from bs4 import BeautifulSoup
import time
# 데이터 관련
import json
import random
import numpy as np
import pandas as pd

# https://kin.naver.com + {step 1.에서 얻은 url} 을 하면 각 QnA로 접속할 수 있다.

df =pd.read_csv('../raw_data/kin.naver.com/qna_links_2021-2025_ver2.csv', encoding='utf-8')
links = df.url.to_list()

url_base = 'https://kin.naver.com'
# headers = { header : '입력!입력!입력!입력!입력!' }

qna_dataset = []
count = 0

# 일단 200개만 해봅시다.. >>> 본격적으로 하면 [:200]만 지웁시다..
for link in  links[:200]:

    count += 1
    url = url_base + link
    response = requests.get(url, headers=headers, timeout=10)
    response.raise_for_status()
    response.encoding = 'utf-8' # utf-8로 인코딩

    bs = BeautifulSoup(response.text, 'html.parser')

    try:
        question = bs.select_one('.questionDetail').text
    except Exception as e:
        print('응답 코드:', response.status_code)
        print(f" !!! 질문 추출 에러: {e} !!!")
        break           # 한번이라도 오류 나면 멈출거라 좋은 선택지는 아니고 continue를 쓰는 게 좋긴 하다.

    try:
        contexts = bs.select('.answerArea .answerDetail .se-viewer')
        content = [ content for content in contexts if 'LG전자고객도우미' in content.text.replace(' ','')][0]
                        # 지식in 특성상, 다른 사람이 답변했을 수 있기 때문에 LG전자 고객도우미 라는 답변이 있는 것만을 가져오는 행위
                        # LG전자고객도우미의 띄어쓰기를 달리할 수 있기 때문에 content.text.replace(' ','')를 해줌으로써 변수 차단
    except Exception as e:
        print('응답 코드:', response.status_code)
        print(f" !!! 답변 추출 에러: {e} !!!")
        break           # 이 break도 마찬가지
    
    question = question.strip()
    answer = content.text.strip()

    qna = {'question': question, 'answer': answer}
    qna_dataset.append(qna)
    
    if count%10 == 0:
        print(f'{count}번째 수행 완료')

    sleep_time = random.uniform(0.7, 1)
    time.sleep(sleep_time)

with open(f'../raw_data/kin.naver.com/lg_electronics_qna_{len(qna_dataset)}.json', 'w', encoding='utf-8') as f:
    json.dump(qna_dataset, f, ensure_ascii=False, indent=4)
```
