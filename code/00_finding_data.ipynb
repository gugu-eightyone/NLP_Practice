### 고려한 자료셋 1. <명일방주 : 엔드필드>의 Google Play 리뷰 크롤링

```python
# !pip install google-play-scraper
```

```python
from google_play_scraper import Sort, reviews

# 명일방주: 엔드필드의 구글플레이 링크는 하단과 같으며, id에 해당하는 부분을 reviews 함수에 입력해주어야한다.
# https://play.google.com/store/apps/details?id=com.gryphline.endfield.gp&hl=ko

result, continuation_token = reviews(
    'com.gryphline.endfield.gp',
    lang='ko',                          # default: 'en'
    country='kr',                       # default: 'us'
    sort=Sort.NEWEST,                   # default: Sort.NEWEST
    count=3,                            # default: 100
    filter_score_with=5                 # default: None(= regardless of the score)
)

result
```

```python
from google_play_scraper import Sort, reviews_all

result_all = reviews_all(
    'com.gryphline.endfield.gp',
    sleep_milliseconds=1,               # default: 0
    lang='ko',                          # default: 'en'
    country='kr',                       # default: 'us'
    sort=Sort.NEWEST,                   # default: Sort.NEWEST
)

type(result_all)
```

```python
import pandas as pd

review_df = pd.DataFrame(result_all)
# review_df.head()

final_df = review_df[['score','content']]
final_df.head()
```

```python
final_df.to_csv('./data/googleplay_reviews/arknights_endfield_googleplay_reviews_20260303.csv', encoding='utf-8', index=False)
```

### 자료셋 2. <원신>의 Google Play 리뷰 크롤링

```python
from google_play_scraper import Sort, reviews_all
import pandas as pd

result_all = reviews_all(
    'com.miHoYo.GenshinImpact',
    sleep_milliseconds=1,               # default: 0
    lang='ko',                          # default: 'en'
    country='kr',                       # default: 'us'
    sort=Sort.NEWEST,                   # default: Sort.NEWEST
)

review_df = pd.DataFrame(result_all)

final_df = review_df[['score','content']]

final_df.to_csv('./data/googleplay_reviews/genshin_googleplay_reviews_20260303.csv', encoding='utf-8', index=False)
```
