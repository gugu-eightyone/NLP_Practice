### LG전자 공식 지식in QnA 200개 임시 전처리

```python
import numpy as np
import pandas as pd

df = pd.read_json('../raw_data/kin.naver.com/lg_electronics_qna_200.json', encoding = 'utf-8')
df
```

##### Step 1. 텍스트 정규화 (Text Normalization)

(1) 영어 문자 대문자 변환 >>> 대문자로 변환 (LG 상호도 대문자고 대개 제품코드는 대문자로 표기되므로)

```python
df = df.applymap(str.upper)
```

(2) 불필요한 문자 제거 / 변환 + 띄어쓰기 적용

```python
# 한글 띄어쓰기 해주는 PyKoSpacing 패키지 설치 (데이터 중간 중간 띄어쓰기가 안 되어있는 경우가 존재)

# !pip install git+https://github.com/haven-jeon/PyKoSpacing.git
```

```python
import re

def cleansing(text):
    # 특수문자 & 노이즈 제거 (ㅠㅠ, ㅋㅋ 등도 제거)
    text = re.sub(r'[^가-힣a-zA-Z0-9\s\.\?\!]', '', text)
    
    # 연속 공백을 하나의 공백으로
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['question'] = df['question'].apply(cleansing)
df['answer'] = df['answer'].apply(cleansing)

df
```

```python
# 띄어쓰기 적용하려 했는데, 생각보다 question에서의 띄어쓰기 결과값이 좋아보이지는 않음 (보류)

# from pykospacing import Spacing

# spacing = Spacing()

# def pkspacing(text):
#     # Pykospacing을 통한 띄어쓰기 적용
#     text = spacing(text)
#     return text

# df['question'] = df['question'].apply(pkspacing)
# df['answer'] = df['answer'].apply(pkspacing)

# df
```

```python
# 답변의 인삿말 제거
df['answer'] = df['answer'].str.split('내용 보았습니다.').str[1]

# 답변의 메크로식 안내문 부분 제거 (답변 맨 뒷부분)
df['answer'] = df['answer'].str.split('감사합니다.').str[0]
```
