##### Step 1. 텍스트 정규화 (Text Normalization)

(1) 영어 문자 대문자 변환 >>> 대문자로 변환 (LG 상호도 대문자고 대개 제품코드는 대문자로 표기되므로)

```python
df1 = df.applymap(str.upper)
```

(2) 불필요한 문자 제거 / 변환 + 띄어쓰기 적용

```python
import re

def cleansing(text):
    # 특수문자 & 노이즈 제거 (ㅠㅠ, ㅋㅋ 등도 제거)
    text = re.sub(r'[^가-힣a-zA-Z0-9\s\.\?\!]', '', text)
    
    # 연속 공백을 하나의 공백으로
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df1['question'] = df1['question'].apply(cleansing)
df1['answer'] = df1['answer'].apply(cleansing)

df1
```

```python
# 위에서 하지 못한 전처리가 하나 존재
# 답변을 보면 링크로 대체한 경우가 종종 존재하는데, 링크 참고해주세요~ 하는 문장들의 세부 표현이 다 다른 듯하다.
# 그래서 '답변'을 문장 토큰화한 후, 각 문장 토큰 list 중에서 링크라는 단어를 포함한 요소부터 그 이후 것들을 다 제거하는 방식으로 하려 했음.

import ast

df2_5 = pd.read_csv('../data/raw_data/milestones/kin.naver.com_step2_milestone.csv', encoding = 'utf-8')

df2_5['answer'] = df2_5['answer'].apply(ast.literal_eval)   # read_csv로 불러왔기에 자료형이 string꼴이 됨. >>> list 형태로 다시 복구해야...
                                                            # KSS로 토큰화한 걸 그대로 사용했다면 이런 작업은 필요 없었겠죠...

def finding_useless(a_list):
    get_rid_of = ['링크', 'HTTPS', 'WWW']

    for index, factor in enumerate(a_list):
        if any(word in factor for word in get_rid_of):
            return a_list[:index]    

    return a_list

df2_5['answer'] = df2_5['answer'].apply(finding_useless)

# df2_5.to_csv('../data/raw_data/milestones/kin.naver.com_step2_5_milestone.csv', encoding = 'utf-8', index = False)
```

```python
# 위에 코드를 돌리니 또 보이는 문제가 발생.
# 이후 궁금한 점이 있으면 고객 상담실로 연락 달라는 문구가 있는 답변들이 다수 존재 >>> 불필요할 듯
# 다만 가끔 고객상담실이라 적을 때도 있고 고객 상담실이라 적을 때도 있음. >>> 크롤링할 때 LG 답변만을 골랐던 방식을 채용하면 될 듯.

def finding_cs(a_list):
    for index, factor in enumerate(a_list):
        if '고객상담실' in factor.replace(' ',''):
            return a_list[:index]    

    return a_list

df2_5['answer'] = df2_5['answer'].apply(finding_cs)

df2_5.to_csv('../data/preprocessed_data/kin.naver.com/preprocessing_200data_step2.csv', encoding = 'utf-8', index = False)
```
